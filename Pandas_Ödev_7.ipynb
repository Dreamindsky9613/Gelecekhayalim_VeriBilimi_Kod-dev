##### 4 Aşağıdaki işlemleri dataframe üzerinde uygulayalım.
- Giyim kategorisinde fiyatı 300'den fazla olan ürünleri gösterin
- Ayakkabı kategorisinde fiyatı 600'den az olan ürünleri gösterin
- Aksesuar kategorisinde fiyatı 100'den fazla olan aksesuarı gösterin

```python
sozluk=[{"Fiyat":[300,180,450,50,700,400,150,80,850,900]}]
fiyat_listesi=sozluk[0]["Fiyat"]
filtreli_fiyat=[fiyat for fiyat in fiyat_listesi if fiyat >300]
filtreli_fiyat
```

```python
import pandas as pd

veri1= pd.DataFrame({"Kategori":["Giyim","Giyim","Giyim","Giyim"],"Ürün":["Kazak","T-shirt","Pantolon","Ceket"],"Fiyat":[300,180,400,900]})
df=pd.DataFrame(veri1)
giyim_filtre=(df["Fiyat"]>300)
giyim_sonuç= df[giyim_filtre]
print(giyim_sonuç)
```

```python
import pandas as pd
veri2=pd.DataFrame({"Kategori":["Ayakkabı","Ayakkabı","Ayakkabı"],"Ürün":["Sandalet","Spor Ayakkabı","Çizme"],"Fiyat":[450,700,850]})
df=pd.DataFrame(veri2)
ayakkabı_filtre=(df["Fiyat"]<600)
ayakkabı_sonuç=df[ayakkabı_filtre]
print(ayakkabı_sonuç)
```

```python
veri3=pd.DataFrame({"Kategori":["Aksesuar","Aksesuar","Aksesuar"],"Ürün":["Küpe","Kolye","Yüzük"],"Fiyat":[50,150,80]})
df=pd.DataFrame(veri3)
aksesuar_filtre=(df["Fiyat"]>100)
aksesuar_sonuç=df[aksesuar_filtre]
print(aksesuar_sonuç)
```
